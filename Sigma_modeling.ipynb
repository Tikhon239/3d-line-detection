```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def gaussian_function_xx(x, sigma):
    var = sigma*sigma;
    return (x*x-var)*np.exp(-x*x/(2*var))/(np.sqrt(2*np.pi*var)*var**2);
```

```python
data = np.array([[gaussian_function_xx(i-10, float(j)/2) for i in range(21)] for j in range(1, 11)])
len(data)
```

```python
plt.figure(figsize=(20, 10))
plt.plot(range(-10, 11), data[0])
```

```python
plt.figure(figsize=(20, 10))
plt.plot(range(-10, 11), data[1])
```
